## Split Data for modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import pickle
x_train, x_test, y_train, y_test = train_test_split(df.drop('Dataset', axis=1), df['Dataset'], test_size=0.25, random_state=123)

print('Train values shape:', x_train.shape)
print('Test values shape:', x_test.shape)
print('Train target shape:', y_train.shape)
print('Test target shape:', y_test.shape)
```

```python
x = df.iloc[:,:-1]
x.head()
```

```python
y = df.iloc[:,-1:]
y.head()
```

```python
def algo_model(x_train, x_test, y_train, y_test, model):
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    print('Prediction')
    print(pred)
    print('-----------------------------------------')
    print("Training score: ",model.score(x_train, y_train))
    print("Testing Score: ", model.score(x_test, y_test))
    print('Confusion Matrix')
    confusion_matrix1 = confusion_matrix(y_test,pred)
    print(confusion_matrix1)
    print('Classification Report')
    print(classification_report(y_test, pred))
    print('-----------------------------------------')
    print('Precision: %.3f' % precision_score(y_test, pred))
    print('Recall: %.3f' % recall_score(y_test, pred))
    print('Accuracy: %.3f' % accuracy_score(y_test, pred))
    print('F1 Score: %.3f' % f1_score(y_test, pred))
    miscal = (1 - accuracy_score(y_test, pred))*100
    print('Miscalculation Rate: %.3f' % miscal, "%")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_1 = GradientBoostingClassifier(random_state=2)
algo_model(x_train, x_test, y_train, y_test, model_1)
```

```python
from sklearn.svm import SVC
model_2 = SVC(kernel = 'linear', C = 100)
algo_model(x_train, x_test, y_train, y_test, model_2)
```

```python
from sklearn.neural_network import MLPClassifier
model_3 = MLPClassifier(random_state=42)
algo_model(x_train, x_test, y_train, y_test, model_3)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,x,y,cv=10)
    knn_scores.append(score.mean())
```

```python
import matplotlib.pyplot as plt
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 4)
algo_model(x_train, x_test, y_train, y_test, knn_classifier)
```
